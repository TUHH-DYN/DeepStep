```python
import json
import os
import sys
import matplotlib.pyplot as plt

colorset = {3: ('#e41a1c', '#377eb8', '#4daf4a'), 4: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3'), 5: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00'), 6: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33'), 7: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628'), 8: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf'), 9: ('#e41a1c', '#377eb8', '#4daf4a', '#984ea3', '#ff7f00', '#ffff33', '#a65628', '#f781bf', '#999999')}

# OPTIONS
# OPTIONS
# OPTIONS

line_styles = []
line_styles.append('-')
line_styles.append('--')
# line_styles.append('--')
# line_styles.append('--')

error_metrics = []
error_metrics.append('loss')
error_metrics.append('val_loss')
# error_metrics.append('domain_mae')
# error_metrics.append('val_domain_mae')
# error_metrics.append('domain_mse')
# error_metrics.append('val_domain_mse')
# error_metrics.append('domain_mes')
# error_metrics.append('val_domain_mes')

name = '$U^p$-Net 64'
# name = '$U^p$-Net x2'
# name = '$U$-Net 96'
# name = '$U$-Net x2'
# name = '$U$-Net noskip 106'
# name = '$U$-Net noskip x2'

training_file = 'model_u_p_net_64_after_350epochs/model_u_p_net_64'
# training_file = 'model_u_p_net_x2_after_350epochs/model_u_p_net_x2'
# training_file = 'model_u_net_96_after_350epochs/model_u_net_96'
# training_file = 'model_u_net_x2_after_350epochs/model_u_net_x2'
# training_file = 'model_u_net_106_noskip_after_350epochs/model_u_net_106_noskip'
# training_file = 'model_u_net_x2_noskip_after_350epochs/model_u_net_x2_noskip'

start_epoch = 0
end_epoch = 350 # max: 350

# END: OPTIONS
# END: OPTIONS
# END: OPTIONS

path_name = os.path.dirname(sys.argv[0])   
abs_path = os.path.abspath(path_name) 
path = abs_path+'/'


notebook_path = os.path.abspath("notebook_visualize_training_evolution.ipynb")

# path = os.path.join(os.path.dirname(notebook_path), "data/model_waveequation")
path = os.path.join(os.path.dirname(notebook_path), "data/model_heatequation")

with open(path+"/"+training_file+"_history.json") as json_file:
    history = json.load(json_file)
    history['epoch'] = range(1,len(history['loss'])+1)

    title = name + ' [' + ', '.join(error_metrics) + ']'
    plt.figure(title, figsize=(12,9)) #, dpi=80

    j = 0
    legend_entries = []
    for error_metric in error_metrics:
        plt.plot(history['epoch'][start_epoch:end_epoch], history[error_metric][start_epoch:end_epoch], line_styles[j], color=colorset[6][j])
        legend_entries.append(error_metric)
        j += 1

    plt.title(title)

    plt.legend(legend_entries, loc='upper right')

    plt.xlabel('epoch')        
    plt.ylabel('loss')

    plt.yscale('log',base=10) 

    # ax = plt.gca()
    # ax.set_ylim(1e-6, 8e-4) 

    plt.show()
```
